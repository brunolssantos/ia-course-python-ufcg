# Atividade sobre validação cruzada

Para esta atividade, você deve implementar o conceito básico por trás da validação cruzada.

Para isso, você deve criar uma função que, dada uma lista com os índices das linhas de um dataframe e o número de folds (pastas), retorne os subconjuntos para aplicar a validação cruzada.

```
def get_folds(indices, n_folds):
    size = len(indices)
    subset_size = round(size / n_folds)
    subsets = [indices[x:x+subset_size] for x in range(0, len(indices), subset_size)]
    conjunto = []
    for i in range(n_folds):
        test = subsets[i]
        train = []
        for subset in subsets:
            if subset != test:
                train = train + subset
        conjunto.append([train, test])
    return conjunto

get_folds(indices, 2) 
```

Após criar sua função, utilize o código abaixo para verificar se sua função está executando corretamente. 

Note que o código abaixo foi constrído pensando que o retorno da função get_folds é uma lista de tuplas, onde o primeiro elemento da tupla é o conjunto com os índices de treino e o segundo o conjunto com os ínidces de teste.

```
import pandas as pd

df = pd.DataFrame({"col1": range(20), "col2": map(lambda n: chr(n), range(97, 117))})

df
```

```
def verifica_conjuntos_cv(df, conjuntos):
    for i in range(len(conjuntos)):
        ind_treino = conjuntos[i][0]
        ind_teste = conjuntos[i][1]
        
        print("Conjunto {}:".format(i + 1))
        
        print("Treino:")
        print(df.iloc[ind_treino])
        
        print("Teste:")
        print(df.iloc[ind_teste])
```

```
indices = list(df.index)

# 2 folds
verifica_conjuntos_cv(df, get_folds(indices, 2))
```

```
# 5 folds
verifica_conjuntos_cv(df, get_folds(indices, 5))
```
