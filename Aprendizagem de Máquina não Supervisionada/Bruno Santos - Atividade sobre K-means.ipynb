```
class K_Means:

  def __init__(self, k=2, tol=0.001, max_iter=300):
    self.k = k
    self.tol = tol
    self.max_iter = max_iter 
  
  def sse(self):
    for centroid in self.centroids:
      for classification in self.classifications:
        sse = np.linalg.norm(self.centroids[centroid]-self.classifications[classification])**2
    return sse

  def fit(self,data):

    self.centroids={}

    for i in range(self.k):
      self.centroids[i]=data[i]

    for i in range(self.max_iter):
      self.classifications = {}

      for i in range(self.k):
        self.classifications[i] = []

      for instance in data:
        distances = [np.linalg.norm(instance-self.centroids[centroid]) for centroid in self.centroids]
        classification = distances.index(min(distances))
        self.classifications[classification].append(instance)

      prev_centroids = dict(self.centroids)

      for i in self.classifications:
        self.centroids[i] = np.average(self.classifications[i], axis=0)

        optimized = True

      for c in self.centroids:
        original_centroid = prev_centroids[c]
        current_centroid = self.centroids[c]
        if (np.linalg.norm(current_centroid - original_centroid)) > self.tol:
          optimized = False
          
      if optimized:
        break
```

```
clusters_num = [2,3,4,5,6]
sse_dic = {}
np.random.shuffle(X)
for i in clusters_num:
  clf = K_Means(k=i, tol=0)
  clf.fit(X)

  colors = 10*["g","r","c","b","k"]

  for centroid in clf.centroids:
    plt.scatter(clf.centroids[centroid][0],clf.centroids[centroid][1], marker="o", color="k", s=150, linewidths=5)

  for classification in clf.classifications:
    color = colors[classification]
    for instance in clf.classifications[classification]:
      plt.scatter(instance[0], instance[1], marker="x", color=color, s=150, linewidths=5)
  title = str(i) + ' cluster(s)'
  plt.title(title)
  plt.show()
  sse_dic[i] = clf.sse()
  print(clf.sse())
```
