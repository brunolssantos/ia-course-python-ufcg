# Artistas que mais venderam discos no Brasil

Este relatório é um exemplo de análise de dados com python. Analisaremos os dados [nessa lista de artistas da Wikipedia](https://pt.wikipedia.org/wiki/Lista_de_recordistas_de_vendas_de_discos_no_Brasil). Os dados foram compilados para o curso AIT 2019.



```
import pandas as pd
import seaborn as sns

sns.set(rc={'figure.figsize':(12,8)})
```

Os dados são lidos diretamente de um site criado por Orion Winter.

```
vendas_artistas = pd.read_csv("https://orionwinter.github.io/datasets/vendas_artistas_br.csv")
```

## Os top 10 que mais venderam

Os 10 artistas que mais venderam discos no Brasil segundo a página da WIkipédia estão listados abaixo. Os artistas que mais venderam são artistas muito famosos (Roberto Carlos e Nelson Gonçalves), mas a lista inclui artistas de sucesso menos perene como Rita Lee ou artistas regionais, como Tonico & Tinoco.

[Bruno] Outra conclusão a que podemos chegar é que artistas mais antigos começaram a carreira bem antes da era do streaming e por isso venderam mais discos do que os artistas recentes em virtude da mudança tecnologica que ocorreu nos últimos anos. Atualmente é muito mais comum consumir músicas por streaming do que comprando mídia física.

```
vendas_artistas.sort_values(by = "vendas", ascending= False).head(n = 10)
```

```
top10 = vendas_artistas.sort_values(by = "vendas", ascending=False).head(n = 10)
```

```
sns.barplot(x = "vendas", y = "artista", data = top10)
```

## Há artistas antigos na lista?

A lista dos que mais venderam inclui alguns (poucos) artistas cujas carreiras começaram há muito. 

[Bruno] Como a lista é sobre os artistas que mais venderam discos, e considerando que a popularização de venda de mídias é um fenomeno mais recente do que o inicio da carreira desses artistas, faz sentido que poucos tenham esse destaque.

```
vendas_artistas.sort_values(by = "inicio_carreira", ascending=True).head(n = 10)
```

## Vendas por gênero musical

Abaixo estão as vendas de todos os artistas, por gênero musical no qual foram classificados na Wikipedia. Os dois artistas mais populares são da MPB, e a MPB tem tanto esses artistas quanto artistas que venderam pouco comparativamente. 

Alguns gêneros, embora tenham vários artistas na lista da Wikipedia, não têm nenhum grande sucesso, como o Pop rock, o Samba e a Música Cristân contemporânea.

```
sns.scatterplot(x = "vendas", y = "genero", hue = "inicio_carreira", data = vendas_artistas, s = 120)
```

## Concentração de Artistas que mais venderam discos entre 1970 e os anos 2000
[Bruno] Aparentemente houve uma popularização da venda de midias fisicas durante a decada de 1970 e o conhecido declinio das mesmas nos anos 2000 quando MP3 começou a se popularizar junto com a popularização da internet.

```
sns.scatterplot(x = "inicio_carreira",  y = "vendas", s = 120, data = vendas_artistas)
```
