# Avaliações de episódios de Game of Thrones e Breaking Bad no IMBD

```
import pandas as pd
import seaborn as sns

sns.set(rc={'figure.figsize':(12,8)})
```

Usamos dados vindos do IMDB e disponíveis no csv desse endereço:

```
dados = pd.read_csv('https://orionwinter.github.io/datasets/series_imdb.csv')
```

```
dados.head(n = 12)
```

```
dados.info()
```

## Avaliação dos episódios de GoT no IMDB

```
got = dados.query('series_name == "Game of Thrones"')
got.head()
```

### Distribuição das avaliações para todos os episódios

```
sns.catplot(y = 'user_rating', data = got, s = 10, kind = "swarm")
```

Olhando os 67 episódios nos dados, vemos que eles têm avaliação entre 8.1 e 9.9. Há um episódio que tem avaliação consideravelmente pior que os demais, enquanto há 4 que têm nota 9.9. Há um número considerável de episódios com avaliação entre 8.5 e 9.3. Outro grupo menor de episódios têm avaliação entre 9.3 e 9.9.

### Apenas algumas temporadas

Observando primeiro apenas a primeira temporada.

```
got_se1 = got.query('season == 1') 
```

```
sns.catplot(y = 'user_rating', data = got_se1, s = 10, kind = "swarm")
```

Observando os 10 episódios da primeira temporada de GoT, vemos que todos os episódios têm avaliação superior a 8.6, e o melhor avaliado tem avaliação em torno de 9.6. No geral, essa é uma temporada bem avaliada. 

Há dois episódios melhor avaliados que os demais, com nota acima de 9.4, depois 5 episódios com nota entre 9 e 9.3, e por fim há 3 episódios com notas de 8.8 ou menos.

Comparando primeiro 1a e 5a temporadas de GoT:

```
got_comparacao_1_5 = got.query('season == 1 | season == 5') 

sns.catplot(x = "season", y = "user_rating", data = got_comparacao_1_5, s = 10, kind = "swarm")
```

A amplitude da distribuição de valores na 5a temporada é maior que na primeira. Há uma maior diferença naquela temporada entre o pior e o melhor episódios. 

O melhor episódio da 5a temporada foi melhor avaliado do que o melhor episódio da 1a temporada. Contudo, o prio episódio da 5a temporada foi pior avaliado que o pior episódio da 1a temporada. Mais ainda, há 4 episódios da 5a temporada que são piores que o pior da 1a temporada. 

```
got_comparacao_1_4 = got.query('season == 1 | season == 4') 

sns.catplot(x = "season", y = "user_rating", data = got_comparacao_1_4, s = 10, kind = "swarm").set(ylim = (8, 10))
```

##Análise da Série Breaking Bad

70% dos 10 episódios melhores avaliados são da temporada 5.

```
bb = dados.query('series_name == "Breaking Bad"')
bb.sort_values(by = "user_rating", ascending =False).head(10)
```

As avaliações dos episódios tem uma tendência de aumento, mostrando que na opnião dos usuários a série melhorou com o tempo na visão dos usuários segundo o IMDb. É possível observar que a mediana aumenta de uma temporada para a outra, exceto pela temporada 3.

```
sns.boxplot(x = "season", y = "user_rating", data = bb, color = "lightblue")
```

A amplitude dos dados da primeira, segunda e quinta temporada são similares, assim como a amplitudade dos dados da terceira e quarta temporada, apesar dos valores mínimo e máximo diferirem apesar dos valores mínimo e máximo da primeira e segunda temporada serem bem similares.

```
sns.catplot(x = "season", y = "user_rating", data = bb, kind = "swarm", s = 10, height = 6, aspect = 1.75, color = "coral")
```

```
sns.swarmplot(x = "season", y = "user_rating", data = bb, color = "white")
sns.violinplot(x = "season", y = "user_rating", data = bb, inner = "box", color = "lightblue")
```

50% dos votos ficaram entre 7,5 e 8,7. Os outros 50% ficaram entre 8,8 e 9,9.

```
sns.swarmplot(x = "user_rating", y = "series_name", data = bb, color = "lightgrey", s = 8)
sns.violinplot(x = "user_rating", y = "series_name", data = bb, inner = "quartile")
```

A maioria dos episódios tiveram nota entre 8,5 e 9,5.

```
sns.distplot(bb['user_rating'], kde = True, bins = 10)
```

Houve um pico de votos nos últimos episódios da última temporada, provavelmente em virtude do desfexo da história que motivos os fãs a votarem mais.  Quanto aos outros episódios, a quantidade de votos é bem similar, ficando abaixo de 20k votos.

```
sns.relplot(x = "series_ep", y = "user_votes", hue = "season", marker = "o", data = bb, height = 4, aspect = 2.5, kind = "line")
```

A quantidade de votos se manteve estável durante todos os episódios da serie, com poucos casos acima de 20k votos.

```
sns.catplot(x = 'season', y = 'user_votes', data = bb, s = 10, kind = "swarm", height = 4, aspect=2.5, color = "purple")
```
