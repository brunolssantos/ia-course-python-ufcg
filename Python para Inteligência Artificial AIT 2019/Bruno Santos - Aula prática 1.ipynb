```
clima = pd.read_csv("https://orionwinter.github.io/datasets/clima_cg_jp_sp-semanal.csv")
```

```
clima.head()
```

```
clima['data'] = pd.to_datetime(clima['semana']).dt.date
```

```
clima.info()
```

```
len(clima["ano"].unique()) #len vem length
```

##Análise Descritiva e Exploratória (EDA)



*   Quais cidades mais quentes?
*   Qual a relação entre os meses mais frios e quentes do ano?
*   A relação entre meses e temperatura é a mesma nas cidades?
*   Será que São Paulo ficou mais quente ao longo dos anos?






```
sns.catplot(x = 'cidade', y = 'tmedia', data = clima, kind = 'violin', aspect = 3 )
```

Comparação da distribuição da tmedia nas três cidades

```
sns.catplot(x = 'mes', y = 'chuva', hue = "cidade", data = clima, kind = 'swarm', dodge = True, aspect = 3.5 )
```

```
clima['tmedia'].max()
```

```
sns.relplot(x = 'data',
            y = 'tmedia',
          hue = 'cidade',
          data = clima,
          kind = 'line',
          aspect = 3.5)
```

```
sns.catplot(x = 'cidade',
            y = 'chuva',
          data = clima,
          kind = 'box',
          aspect = 2)
```

```
sns.catplot(x = 'mes',
            y = 'chuva',
          hue = 'cidade',
          data = clima,
          kind = 'box',
          dodge = True,
          aspect = 3.5)
```

```
sns.relplot(x = 'data',
            y = 'chuva',
          hue = 'cidade',
          data = clima.query('ano > 2015'),
          kind = 'line',
          row = 'cidade',
          height = 3,
          aspect = 3.5)
```

##E a relação entre temperatura e chuva

```
sns.relplot(x = 'tmedia',
           y = 'chuva',
           hue = 'cidade',
           data = clima,
           kind = 'scatter',
           height = 4,
           col = 'cidade',
           alpha = 0.5)
```
