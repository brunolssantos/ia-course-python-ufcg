# Artistas que mais venderam discos no Brasil

Este relatório é um exemplo de análise de dados com python. Analisaremos os dados [nessa lista de artistas da Wikipedia](https://pt.wikipedia.org/wiki/Lista_de_recordistas_de_vendas_de_discos_no_Brasil).

```
import pandas as pd
import seaborn as sns

sns.set(rc={'figure.figsize':(12,8)})
```

```
vendas_artistas = pd.read_csv("https://orionwinter.github.io/datasets/vendas_artistas_br.csv")
```

```
vendas_artistas.sort_values(by = "vendas", ascending = True)
```

```
vendas_artistas.sort_values(by = "vendas", ascending= False).head(n = 10)
```

```
vendas_artistas.sort_values(by = "inicio_carreira", ascending=True).head(n = 10)
```

## Agora com visualizações

```
top10 = vendas_artistas.sort_values(by = "vendas", ascending=False).head(n = 10)
```

```
sns.barplot(x = "vendas", y = "artista", data = top10)
```

```
sns.scatterplot(x = "vendas", y = "artista", data = top10)
```

```
sns.scatterplot(x = "vendas", y = "genero", data = vendas_artistas, s = 120)
```

```
sns.scatterplot(x = "vendas", y = "genero", hue = "inicio_carreira", data = vendas_artistas, s = 120)
```

```
sns.scatterplot(x = "inicio_carreira",  y = "vendas", s = 120, data = vendas_artistas)
```

A maioria dos artistas que começaram a carreira mais recentemente venderam menos discos do que os que começaram antes. Isso pode levar a crer que com mais tempo de carreira, o artista teve mais tempo de vender discos.
